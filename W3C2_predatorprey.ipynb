```python
 # Modeling population Dynamics: Predator Prey Model AKA Lotka-Volterra
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps =20000
a = 2     #limited grow rate of bunnies
b = 1.2
c = 0.5
d = 1
h= 3    #rate of hunters killing foxes   #new element #how do i make iit so that the hunters kill foxes and bunnies? 

dt = 0.001
possible_ds = [0.5,1,1.5,2]

x = np.zeros(nsteps)
y = np.zeros(nsteps)

fig1 = plt.figure(figsize=(16,5));
fig2 = plt.figure(figsize=(5,5));
ax2= plt.axes()

for itrail in range(len(possible_ds)):
    d = possible_ds[itrail]
    x[0] = 10
    y[0] = 1

    for t in range(1, nsteps):
        x[t] = x[t-1] + dt*(a*x[t-1] - b*x[t-1]*y[t-1])
        y[t] = y[t-1] + dt*(c*x[t-1]*y[t-1]-d*y[t-1])
    
    ax1 = fig1.add_subplot(1,len(possible_ds),itrail+1) 
    ax1.plot(x, label = 'Bunnies')
    ax1.plot(y, label = 'Foxes')
    ax1.legend(loc='upper right')

    ax2.plot(x,y,label=d)
    plt.xlabel('Bunnies')
    plt.ylabel('Foxes')
    ax2.legend(loc='upper right')
```
