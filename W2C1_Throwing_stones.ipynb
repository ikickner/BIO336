```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Attempt #1: throwing the step upwards
a = -1      #acceleration; neg bc it tries to reverse the flight of stone
v = 5       #we threw the stone upward, so its positive
y = 0       #we threw it from the ground, so y=0
x = 0       #we throw it from our house and see how far it goes 
dt =0.01    #our step, numerical integration step, is supposed to be small
insteps = 1000
history = np.zeros(insteps)    #we will plot this eventually
for i in range(insteps):
    v = v + a*dt
    x = x + v*dt
    history[i] = x
    
plt.plot(history)
```

```python
# Attempt #2: throwing it at an angle
a = -1
vy = 5
vx = 1
x = 0
y = 0
dt = 0.01
insteps = round(10/dt)

for i in range(insteps):
    vy = vy + a*dt         #update vertical speed
    
    x = x +vx*dt           #update x position
    y = max(y + vy*dt, 0)  #update y position
    
    plt.plot(x,y,'k.');    #just plot a new point, instead of remembering x and y, bc we're lazy rn
```

```python
# Attempt 3: add friction
a = -1
vy = 5
vx = 1
x = 0
y = 0
f = 0.1           # how important friction is relative to gravity 
dt = 0.05
insteps = round(10/dt)
    
for i in range(insteps):
        vy = (vy + a*dt) - (vy*f)
        vx = vx - vx*f
        
        x = (x + vx*dt)
        y= max(y + vy*dt, 0)
        
        plt.plot(x,y,'k.');
```
